```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

### Removing the unused or irrelevant columns

```python
to_drop = ['cp',
           'fbs', 
           'restecg', 
           'thalach',
           'exang', 
           'oldpeak', 
           'slope', 
           'ca', 
           'thal']
df.drop(to_drop, axis=1, inplace=True)
df.head(5)
```

### Renaming the column names

```python
new_names = {
    'age': 'Age',
    'sex': 'Sex',
    'trestbps': 'Resting_BP',
    'chol': 'Cholesterol',
    'target': 'Heart_Disease'
}
df.rename(columns=new_names, inplace=True)
df.head()
```

### Replacing the value of the rows

```python
df['Sex'] = df['Sex'].replace({1: 'Male', 0: 'Female'})

df.head()
```

```python
df.to_csv('cleaned_heart_data.csv', index=False)
print("Data Cleaning Successfully Completed and Saved!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Distribution of Male and Female

```python
#This chart shows the number of males and females in our dataset. We can see that there are more male records compared to females
sns.countplot(x='Sex', data=df,hue='Sex', palette='viridis',legend=False)

plt.title('Distribution of Male and Female')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Age vs Cholesterol Level (Scatter Plot)

```python
#This scatter plot shows the relationship between age and cholesterol. Generally, as age increases, cholesterol levels also tend to rise for many individuals.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Cholesterol', hue='Sex', data=df)

plt.title('Age vs Cholesterol Level')
plt.show()
```

### Age Distribution (Histogram)

```python
#This histogram shows that most of the people in this study are between the ages of 50 and 60. The data is mostly concentrated in middle-aged groups
sns.histplot(df['Age'], bins=20, kde=True, color='blue')

plt.title('Age Distribution')
plt.show()
```
